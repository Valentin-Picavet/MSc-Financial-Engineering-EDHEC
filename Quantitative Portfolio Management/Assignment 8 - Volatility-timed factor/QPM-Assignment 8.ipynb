# Question 1

Please use mean-variance optimization to combine :
- The original (without timing) factor, $f_{t+1}$;
- The volatility-managed version of this factor, $f^{\sigma}_{t+1}$

```python
from scipy.optimize import minimize

def mean_variance_objective(weights, returns):
    portfolio_return = weights.T @ returns.mean()
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov(), weights)))
    return -portfolio_return / portfolio_volatility

def maximize_SR(returns):
    num_assets = returns.shape[1]

    # Set initial weights
    initial_weights = np.ones(num_assets) / num_assets
    # Minimize the negative SR  to maximize the SR 
    result = minimize(mean_variance_objective, initial_weights, args=(returns,), method='SLSQP')

    return result.x
```

```python
mkt=pd.concat([data_volatility_managed['Market'],data_comparison['Market']],axis=1)
optimal_mkt_weight=maximize_SR(mkt)
optimal_mkt_weight
```

To overcome these limitations, investors should turn their attention to the results derived from multifactor portfolios by DeMiguel, Martin-Utrera, and Uppal. These findings indicate substantial performance enhancements resulting from volatility timing, encompassing aspects such as out-of-sample performance, net transaction costs, and applicability in both low- and high-sentiment periods.

While the analysis suggests an augmented Sharpe ratio for portfolios incorporating the volatility-timed factor, it is crucial to recognize various limitations. Historical performance might not reliably predict future outcomes. Implementing volatility-timing strategies in practical scenarios introduces challenges such as transaction costs, liquidity constraints, and the potential for model overfitting.
