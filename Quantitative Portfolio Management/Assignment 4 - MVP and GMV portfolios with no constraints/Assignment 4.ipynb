# Question 0:

- From the data that we used for the previous assignment, select the following 10 companies (these are the first 10 companies with no missing data): ”MMM”,”AOS”,”ABT”,”ADM”,”ADBE”,”ADP”,”AES”,”AFL”,”A”,”AKAM”
- So, our “new” dataset for this assignment will consist of monthly returns you had computed in the last assignment, but just for these 10 companies.
- To reduce the work required for this assignment, please assume that the risk-free rate of return is zero.

```python
L_stock=['MMM','AOS','ABT','ADM','ADBE','ADP','AES','AFL','A','AKAM']
for stock in stocks_prices.columns:
    if stock not in L_stock:
        stocks_prices= stocks_prices.drop(stock, axis=1)

stocks_prices
```

- We compute the daily log return

```python
log_return_df=np.log(stocks_prices / stocks_prices.shift())
log_return_df
```

```python
L_year =[]
L_month=[]
for date in log_return_df.index:
    L_year.append(date.year)
    L_month.append(date.month)
log_return_df['Year']=L_year
log_return_df['Month']=L_month
log_return_df
```

- Calculation of monthly returns

```python
L_year_unique=list(np.unique(L_year))
L_month_unique=list(np.unique(L_month))
L_index=[]

for year in L_year_unique:
    for month in L_month_unique:
        L_index.append(str(month) + "/" + str(year))
        
df_monthly_return=pd.DataFrame(index=L_index, columns=log_return_df.columns[:10])
for stock in log_return_df.columns[:10]:
    L_monthly=[]
    for year in L_year_unique:
        for month in L_month_unique:
            data=log_return_df[(log_return_df['Year']== year) & (log_return_df['Month']== month)][stock]
            m=data.mean()*21 # 21 trading days per month
            L_monthly.append(m)
    df_monthly_return[stock]=L_monthly
```

# Question 1:

- Choose the estimation window to be T est = 60 months of monthly returns. Call this the estimation sample. Use the estimation sample to compute the following two portfolio strategies:
     - a. mean-variance portfolio (MVP) without constraints on the size of the weight (assume that a risk-free rate is available, with the risk-free rate equal to zero);
    - b. global minimum variance (GMV) portfolio without constraints on the size of the weight.

```python
import numpy as np
from scipy.optimize import minimize

def objective_function(weights, returns):
    # Calculate mean and standard deviation of portfolio returns
    portfolio_return = np.dot(returns.mean(axis=0), weights)
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov(), weights)))

    # Calculate the negative Sharpe ratio (since we want to maximize it)
    sharpe_ratio = -portfolio_return / portfolio_volatility # as rf=0

    return sharpe_ratio

def maximize_sharpe_ratio(returns):
    num_assets = returns.shape[1]

    # Set initial weights (equally weighted)
    initial_weights = np.ones(num_assets) / num_assets

    # Set optimization constraint (sum of weights equals 1)
    constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})

    # Minimize the negative Sharpe ratio to maximize the Sharpe ratio
    result = minimize(objective_function, initial_weights, args=(returns,), method='SLSQP')

    return result.x

T_est=60
estimation_sample=df_monthly_return[:T_est]
sharpe_max_weights = maximize_sharpe_ratio(estimation_sample)

print("MVP Portfolio weights:")
print(sharpe_max_weights)
```

```python
def gmv(sigma):
    n=len(sigma)
    sigma_inverse=np.linalg.inv(sigma) #calculate a matrix inverse
    one_vector=np.ones(n)
    numerator=np.dot(sigma_inverse, one_vector)
    weights = numerator / np.sum(numerator)
    return weights
```

```python
T_est=60
estimation_sample=df_monthly_return[:T_est]
cov_matrix=estimation_sample.cov()
gmv_weights=gmv(cov_matrix)
```

```python
print("GMV Portfolio Weights:")
print(pd.Series(gmv_weights, index=cov_matrix.index))
```

# Question 2:

- Now use a rolling window of T est = 60 months to estimate the portfolio weights for the two strategies listed above for each of the T − T est months. That is, repeat the calculations of the previous question for all the dates after the first 60 months.

```python
gmv_weights_list = []
T_est=60

for i in range(T_est, len(df_monthly_return)):
    returns_window = df_monthly_return.iloc[i - T_est:i]
    covariance_matrix = returns_window.cov()
    gmv_weights = gmv(covariance_matrix.values)
    gmv_weights_list.append(gmv_weights)

gmv_weights_df = pd.DataFrame(gmv_weights_list, columns=df_monthly_return.columns)

print("GMV Portfolio Weights:")
gmv_weights_df
```

```python
T_est = 60
num_months = len(df_monthly_return)
columns = df_monthly_return.columns

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(index=df_monthly_return.index[T_est:], columns=columns)

# Iterate through each date after the first 60 months
for i in range(T_est, num_months):
    estimation_sample = df_monthly_return[i - T_est:i]
    sharpe_max_weights = maximize_sharpe_ratio(estimation_sample)

    # Store the results in the DataFrame
    result_df.loc[df_monthly_return.index[i]] = sharpe_max_weights

# Display the resulting DataFrame
print("MVP Portfolio Weights:")
result_df
```

# Question 3

- Use the time-series of portfolios weights for each of the two portfolio strategies, to compute the out-of-sample portfolio returns. That is, for each of the two portfolio strategies that you estimate at each date t, compute its out-of-sample return in month t + 1.


```python
T_est = 60
num_months = len(df_monthly_return)
columns = df_monthly_return.columns

result_df = pd.DataFrame(index=df_monthly_return.index[T_est:], columns=columns)

for i in range(T_est, num_months):
    estimation_sample = df_monthly_return[i - T_est:i]
    sharpe_max_weights = maximize_sharpe_ratio(estimation_sample)
    result_df.loc[df_monthly_return.index[i]] = sharpe_max_weights

# Compute out-of-sample portfolio returns
out_of_sample_returns = (df_monthly_return.shift(-1).loc[result_df.index] * result_df).sum(axis=1)

# Create a DataFrame for out-of-sample returns
out_of_sample_returns_mvp = pd.DataFrame(out_of_sample_returns, columns=['Out_of_Sample_Return'])

out_of_sample_returns_mvp[:215]
```

```python
T_est = 60
gmv_weights_list = []
out_of_sample_returns_list = []
for i in range(T_est, len(df_monthly_return)-1):
    returns_window = df_monthly_return.iloc[i - T_est:i]
    covariance_matrix = returns_window.cov()
    gmv_weights = gmv(covariance_matrix.values)
    gmv_weights_list.append(gmv_weights)
    returns_next_month = df_monthly_return.iloc[i + 1]
    out_of_sample_return = np.dot(gmv_weights, returns_next_month)
    out_of_sample_returns_list.append(out_of_sample_return)

out_of_sample_returns_gmv = pd.DataFrame(out_of_sample_returns_list, columns=['Out_of_Sample_Return'])
print("\nOut-of-Sample Portfolio Returns:")
out_of_sample_returns_gmv
```

# Question 4

- Now, compute the Sharpe ratio of the out-of-sample returns for the two portfolio strategies. Which strategy has the higher Sharpe ratio?

```python
sharpe_ratio_mvp = out_of_sample_returns_mvp.mean() / out_of_sample_returns_mvp.std() # assuming rf=0
sharpe_ratio_gmv = out_of_sample_returns_gmv.mean() / out_of_sample_returns_gmv.std()
print("\nSharpe Ratio MVP:")
print(sharpe_ratio_mvp)

print("\nSharpe Ratio GMV:")
print(sharpe_ratio_gmv)
```
