# Question 1

- Explain why one might expect these nine factors to be related to stock returns. Write only a few sentences (2 or 3 sentences) for each factor. (Feel free to use ChatGPT, but reading the original paper would be much more educational.)

- Market (MKT): The Market factor, often represented by the excess return of a broad market index like the S&P 500, is a measure of overall market performance. Stocks are influenced by macroeconomic conditions, corporate earnings, and investor sentiment. Positive economic indicators, strong corporate earnings, and optimistic investor sentiment tend to drive market returns higher. Consequently, individual stock prices are impacted as they are part of the broader market.


- Size (SMB): The Size factor (SMB) is based on the historical observation that small-cap stocks have, on average, outperformed large-cap stocks. This can be attributed to the additional risk associated with smaller companies as well as their potential for higher growth. Small-cap stocks are often considered riskier but may offer greater growth opportunities, leading to a size premium.


- Value (HML): The Value factor (HML) captures the historical outperformance of value stocks over growth stocks. High Minus Low (HML) is a value premium. It represents the spread in returns between companies with a high book-to-market value ratio (value companies) and companies with a low book-to-market value ratio. The HML factor attempts to capture the premium associated with investing in stocks that are considered undervalued by traditional valuation metrics.


- Profitability (RMW): The Profitability factor (RMW) focuses on the historical outperformance of stocks of companies with strong profitability compared to those with weak profitability. Companies with higher profitability, as measured by return on assets or return on equity, are often considered more stable and capable of sustaining their earnings over time. Investors may be willing to pay a premium for such companies.


- Investment (CMA): The Investment factor (CMA) represents the historical outperformance of stocks of companies with conservative investment policies over those with aggressive investment policies. Conservative firms are characterized by more prudent investment strategies, avoiding excessive risk. This factor attempts to capture the premium associated with investing in stocks of companies that focus on long-term stability and value preservation.


- Momentum (UMD): The Momentum factor (UMD) captures the historical outperformance of stocks with strong past performance relative to those with weak past performance. Investors often exhibit a tendency to extrapolate trends in stock prices, leading to the continuation of price movements in the short to medium term. This factor attempts to exploit the momentum effect observed in stock prices.


- Return on Equity (ROE): The Return on Equity factor (ROE) focuses on the historical outperformance of stocks of companies with higher return on equity. ROE is a measure of a company's profitability and efficiency in generating returns for its shareholders. Stocks of companies with consistently high ROE may be perceived as having a competitive advantage and are expected to outperform.


- Investment in Assets (IA): The Investment in Assets factor (IA) considers the historical outperformance of stocks of companies with high investment in assets relative to low investment. This factor attempts to capture the impact of companies making strategic investments in their business operations. Companies investing in assets for future growth may be expected to outperform those with lower investment levels.


- Betting Against Beta (BAB): The Betting Against Beta factor (BAB) represents the historical outperformance of low-beta stocks over high-beta stocks. This factor challenges the traditional risk-return trade-off, suggesting that low-beta stocks (those less sensitive to market movements) can provide higher risk-adjusted returns. Investors may be willing to pay a premium for stocks that exhibit lower systematic risk than predicted by their beta.


In summary, each factor is grounded in economic theories and empirical observations that aim to explain the variations in stock returns based on specific characteristics or strategies.

```python
data['Dates'] = pd.to_datetime(data['Dates'], format='%Y%m')
data = data.set_index('Dates')
data.index = data.index.strftime('%Y-%m')
```

 To find the optimal weights, we use this formula from the course : $\text{maxwMVU} = \left( R_f + w^T(\mu - R_f \mathbf{1}_N) - \frac{\gamma}{2} w^T V w \right)$

```python
import numpy as np
from scipy.optimize import minimize

def objective_function_MVP(weights,returns,gamma):

    portfolio_return = weights.T @ returns.mean() # we assume rf=0
    term_2 = 0.5*gamma*weights.T @ returns.cov() @ weights

    return -(portfolio_return - term_2)

def maximize_utility_function(returns,gamma):
    num_assets = returns.shape[1]

    # Set initial weights (equally weighted)
    initial_weights = np.ones(num_assets) / num_assets

    # Constraint on the mkt weight to be 1
    constraints = ({'type': 'eq', 'fun': lambda weights: weights[0] - 1}) 

    # Minimize the negative uility function to maximize the utility function
    result = minimize(objective_function_MVP, initial_weights, args=(returns,gamma), method='SLSQP', constraints=constraints)

    return result.x

gamma = 5
weights_parametric=maximize_utility_function(data,gamma)

print("Weights:")
print(pd.Series(weights_parametric, index=data.columns))
```

- Find the Sharpe ratio for each of the nine factors and compare it to that of the parametric portfolio you have identified in the previous question.

```python
expected_returns = [data[factor].mean() for factor in data.columns]
expected_returns
```

```python
st_dev = [data[factor].std() for factor in data.columns]
st_dev
```

```python
sharpe_ratio_factors = pd.DataFrame(index=data.columns)
sharpe_ratio_factors['SR']=[(er / sd)*np.sqrt(12) for er, sd in zip(expected_returns, st_dev)]
sharpe_ratio_factors
```

```python
data_weighted=pd.DataFrame(index=data.index, columns=['Parametric Portfolio Returns'])
for i in range (len(data)):
    data_weighted.iloc[i]=(weights_parametric*data.iloc[i]).sum()
    
data_weighted
```

```python
returns_parametric=data_weighted.mean()
print('Returns Parametric Portfolio')
print(returns_parametric.iloc[0])
```

```python
st_dev_parametric = data_weighted.std()
print('St dev Parametric Portfolio')
print(st_dev_parametric.iloc[0])
```

```python
SR_parametric=(returns_parametric/st_dev_parametric)*np.sqrt(12)
print('Sharpe Ratio Parametric Portfolio')
print(SR_parametric.iloc[0])
```

We see that the Sharpe Ratio of the Parametric Portfolio is higher than the Sharpe Ratio of any of the factors.

-  Having obtained the optimal θ vector, please explain how one would obtain the optimal portfolio weights for each of the Nt = 2000 assets that are used to form each of the nine factors.

To obtain the optimal proftolio weights for each of the $N_t$ assets, we use the following formula :
$
w_t(\theta) = w_{b,t} + \frac{F_{1,t}\theta_1 + F_{2,t}\theta_2 + \ldots + F_{K,t}\theta_K}{N_t}
$

where :

$w_{b,t}$ is the benchmark portfolio at time t, determined by the client or investor

$F_{K,t}$ is the long-short characteristic portfolio obtained by standardizing the kth firm-specific characteristic at time t

$\theta_K$ is the weight on the kth characteristic portfolio in the parametric portfolio

$N_t$  is the number of firms at time t

So the only unknown is $\theta$ in this formula, but we have just computed it, so to have the optimal portfolio weights for each assets we just have to use the above formula.

What this formula shows is that the optimal portfolios weights are obtained by adding to the benchmark portfolio a linear combination of long-short characteristic portfolios obtained by standardizing K firm-specific characteristics cross-sectionally.
The weights of the characteristics in the parametric portfolio are scaled by the number of stocks Nt so that they are meaningful for the case with a varying number of stocks.
