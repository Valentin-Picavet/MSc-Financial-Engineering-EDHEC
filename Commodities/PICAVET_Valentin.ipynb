### Question 4
Check the stationarity of the `Heating Oil - Brent` spread. What are your thoughts on the results?

In order to test for stationarity, we need to test for something called a *unit root*. Autoregressive unit root tests are based on the following hypothesis test:

$$
\begin{aligned}
H_{0} & : \phi = 1 \implies y_{t} \sim I(0) | (unit \ root) \\
H_{1} & : |\phi| < 1 \implies y_{t} \sim I(0) | (stationary)  
\end{aligned}
$$

It's referred to as a unit root test because under the null hypothesis, the autoregressive polynomial of $\scr{z}_{t}, \ \phi (\scr{z})= (1-\phi \scr{z})  = 0$, has a root equal to unity. 

$y_{t}$ is trend stationary under the null hypothesis. If $y_{t}$ is then first differenced, it becomes:

$$
\begin{aligned}
\Delta y_{t} & = \delta + \Delta\scr{z}_{t} \\
\Delta \scr_{z} & = \phi\Delta\scr{z}_{t-1} + \varepsilon_{t} - \varepsilon_{t-1} \\
\end{aligned}
.$$

The test statistic is

$$ t_{\phi=1}=\frac{\hat{\phi}-1}{SE(\hat{\phi})}$$

$\hat{\phi}$ is the least square estimate and $SE(\hat{\phi})$ is the usual standard error estimate. The test is a one-sided left tail test. If $y_{t}$ is stationary, then it can be shown that

$$\sqrt{T}(\hat{\phi}-\phi)\xrightarrow[\text{}]{\text{d}}N(0,(1-\phi^{2}))$$

or 

$$\hat{\phi}\overset{\text{A}}{\sim}N\bigg(\phi,\frac{1}{T}(1-\phi^{2}) \bigg)$$

and it follows that $t_{\phi=1}\overset{\text{A}}{\sim}N(0,1)$. However, under the null hypothesis of non-stationarity, the above result gives

$$
\hat{\phi}\overset{\text{A}}{\sim} N(0,1)
$$


To do so, I implement the Augmented Dickey Fuller test using `adfuller`.

```python
from statsmodels.tsa.stattools import adfuller

df_test=adfuller(df_concat['Spread Heating Oil-Brent'])
df_output=pd.Series(df_test[0:4],index=['Test statistic','p-value',"Number of lags used",'Number of observations used'])
critical_values=pd.Series(df_test[4].values(),index=df_test[4].keys())
print(f'Dickey-Fuller Result: \n{df_output} \n\nDickey-Fuller Critical Values: \n{critical_values}' )
```

The Dickey-Fuller p-value is above 1%,5% and 10%. Therefore, we do not reject the null Hypothesis at the 1%, 5% and 10% level. That is the spread is not stationary at the 1%, 5% and 10% level.

This result is right because stock/commodities prices can be thought of being a cumulative sum of mean-independent increments due to economic (and other types of) shocks. This is per definition a process with a unit root:
$
X_t = X_{t-1} + \varepsilon_t = (X_{t-2} + \varepsilon_{t-1}) + \varepsilon_t = \cdots = \sum_{\tau=0}^{t} \varepsilon_{\tau}
$

Meanwhile, the increments $\varepsilon_{\tau}$ are mean-independent and "almost stationary", but their scale grows with the level of the stock/commodity price so it is not stationary.

Now that I've found the HTML code for the table that interests me I need to find the name of the columns. By looking at the above HTML code we notice that they are separated by 'th'.

```python
titles_gas=soup_gas.find_all('th')
table_titles_gas=[title.text for title in titles_gas]
print(table_titles_gas)
```

Now that I've got the name of the different columns, I create an empty data frame with the right columns name.

```python
colonnes_gas=[table_titles_gas[0]]+table_titles_gas[6:]
df_gas=pd.DataFrame(columns=colonnes_gas)
df_gas
```

We also notice that the data that fills the table on the website is separated by 'td' in the HTML code. Thus, I create an empty list that I will store with the data I extracted from the HTML code. I notice that the data in the HTML code is in the good order. I fill my data frame with the data I extracted.

```python
column_data_gas=table_gas.find_all('tr')

filtered_data_gas = []
for row in column_data_gas[2:]:
    row_data=row.find_all('td')
    individual_row_data=[data.text.strip() for data in row_data]
    if individual_row_data != ['']:
        filtered_data_gas.append(individual_row_data)

for data_row in filtered_data_gas:
    df_gas.loc[len(df_gas)]=data_row

df_gas
```

Now that I've reproduced the original table from the website, I want to create a new data frame that I can manipulate more easily because we notice some columns are empty, etc and it could be complicated to manipulate. As I have 5 End Date and Value columns, to differentiate them to then be able to extract the data, I rename them with a number.

```python
nouveaux_noms_colonnes_gas = ['Year-Month'] + [f'End Date{i}' if i % 2 != 0 else f'Value{i//2}' for i in range(1, len(df_gas.columns))]
df_gas.columns = nouveaux_noms_colonnes_gas

df_gas.rename(columns={df_gas.columns[3]: 'End Date2'}, inplace=True)
df_gas.rename(columns={df_gas.columns[5]: 'End Date3'}, inplace=True)
df_gas.rename(columns={df_gas.columns[7]: 'End Date4'}, inplace=True)
df_gas.rename(columns={df_gas.columns[9]: 'End Date5'}, inplace=True)

df_gas
```

Now, I create 2 functions that are going to extract the dates from the different End Date columns and the values corresponding. It will give me 2 lists with the right extracted data.

```python
def concat_date(row):
    year_month = row['Year-Month'][:4]
    end_dates = []
    for i in range(1, 6):
        end_date = row[f'End Date{i}']
        if end_date: 
            end_dates.append(year_month + '-' + end_date)
    return end_dates

def concat_value(row):
    values = []
    for i in range(1, 6):
        value = row[f'Value{i}']
        if value: 
            values.append(value)
    return values
```

I arrange the format of the data extracted because they are object or strings. So for the dates I convert them to date time type and for the values they were string so I convert them to integer.

```python
df_gas['Concatenated Dates'] = df_gas.apply(concat_date, axis=1)
all_dates_gas = [date for sublist in df_gas['Concatenated Dates'] for date in sublist]
all_dates_gas = pd.to_datetime(all_dates_gas)
all_dates_gas
```

```python
df_gas['Concatenated Values'] = df_gas.apply(concat_value, axis=1)
all_values_gas = [value for sublist in df_gas['Concatenated Values'] for value in sublist]
all_values_gas = [value.replace(',', '') for value in all_values_gas]
all_values_gas = list(map(int, all_values_gas))
print(all_values_gas)
```

Then I create the final data frame containing the dates as index and the corresponding values of demand in the first and only column of the data frame. Note that I do not have any order problem as I extracted the data from top to bottom always taking the value corresponding to the right end date. The result is a clean data frame that I can exploit.

```python
gasoline_demand=pd.DataFrame(all_values_gas,index=all_dates_gas,columns=['Weekly U.S. Product Supplied of Finished Motor Gasoline  (Thousand Barrels per Day)'])
gasoline_demand.index.name='Date'
gasoline_demand
```

It seems like there’s a trend here. It appears that the seasonality has a period of one year. Next, I’ll look into a tool I can use to further examine the seasonality and break down our time series into its trend, seasonal, and residual components.

We can see that the amplitudes on the seasonality is larger or smaller depending on the trend. Thus we have a multiplicative seasonality. We can model this with this equation:
Y[t] = T[t] *S[t] *e[t] where T[t] is the trend, S[t] the seasonality and e[t] the residual.

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Perform seasonal decomposition
decomposition = seasonal_decompose(gasoline_demand['Weekly U.S. Product Supplied of Finished Motor Gasoline  (Thousand Barrels per Day)'],
                                    model='multiplicative', period=52)  

plt.figure(figsize=(12, 6))
plt.subplot(411)
plt.plot(gasoline_demand, label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend(loc='upper left')
plt.tight_layout()

# Remove the seasonal component
gasoline_demand_seasonality_removed = gasoline_demand['Weekly U.S. Product Supplied of Finished Motor Gasoline  (Thousand Barrels per Day)']/decomposition.seasonal.dropna()

plt.figure(figsize=(20, 8))
plt.plot(gasoline_demand.index, gasoline_demand,color='red',label='Original')
plt.plot(gasoline_demand_seasonality_removed.index, gasoline_demand_seasonality_removed,color='blue',label='Seasonality Removed')
plt.legend(loc='upper left')
plt.title('Original Data vs Data with Seasonality Removed')
plt.show()
```

By looking at the seasonal component, we notice that there is a peak of demand during the summer which is verified in real life as people are using more their car during the summer holidays than during the winter.

```python
increase=(decomposition.seasonal.max()-decomposition.seasonal.min())/decomposition.seasonal.min()*100
print('Increased Gasoline demand between the winter and summer season is:',increase,"%")
```

We know that PED = $-\frac{dD}{dP} \times \frac{P}{D}$
where D is the demand and P is the price. 

To calculate the price-elasticity of the demand we do a univariate OLS where y is the demand and x is the price such that $y = \beta \cdot x + \alpha$ . 

Then $\frac{{dy}}{{dx}} = \beta$ and we are able to compute the price-elasticity of the demand by multiplying the beta of the regression by the mean of the price/demand.

But first I need to resample my daily prices in weekly prices in order to comply with the weekly demand. Otherwise I will have a problem when running the regression because I will get the error message 'Endog and Exog indices are not aligned'.

I need to do the regression on the returns and the variation of the demand. To do so I `use pct_change`.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

X = df_concat['US Gasoline Prices'].resample('W-Fri').last().iloc[:-1]
X = X.pct_change().dropna()
Y = gasoline_demand_seasonality_removed["2010-01-08":"2024-03-22"].pct_change().dropna()
X = sm.add_constant(X) 
model=sm.OLS(Y,X)
results=model.fit()
results.params

alpha=results.params.values[0]
beta=results.params.values[1]

price_elasticity = - beta * (np.mean(X['US Gasoline Prices']) / np.mean(Y))

print("Price-Elasticity of Demand for Gasoline:", price_elasticity)
```

We have a Price-Elasticity of demand really close to 0 which indicates that the demand for the good is nearly perfectly inelastic as it is the case in practice.

I find parameters a,b and sigma of the Vasicek model using the MLE method.

```python
def VasicekCalibration(margin, dt=1/12):
    n = len(margin)
    
    # Implement MLE to calibrate parameters     
    Sx = sum(margin[0:(n-1)])
    Sy = sum(margin[1:n])
    Sxx = np.dot(margin[0:(n-1)], margin[0:(n-1)])
    Sxy = np.dot(margin[0:(n-1)], margin[1:n])
    Syy = np.dot(margin[1:n], margin[1:n])
    
    b = (Sy * Sxx - Sx * Sxy) / (n * (Sxx - Sxy) - (Sx**2 - Sx*Sy))
    a = -np.log((Sxy - b * Sx - b * Sy + n * b**2) / (Sxx - 2*b*Sx + n*b**2)) / dt
    c = np.exp(-a * dt)
    sigmah2 = (Syy - 2*c*Sxy + c**2 * Sxx - 2*b*(1-c)*(Sy - c*Sx) + n*b**2 * (1-c)**2) / n
    sigma = np.sqrt(sigmah2*2*a / (1-c**2))
    M0 = margin[n-1]
    
    return [a, b, sigma, M0]
```

```python
params = VasicekCalibration(Margin)
a = params[0]
b = params[1]
sigma = params[2]
M0 = params[3]
print(a,b,sigma,M0)
```

Then I use the estimated parameters to model the margin paths using Monte Carlo simulation. I use the closed form solution:

$\text{margin}_{t_i} = \text{margin}_{t_{i-1}} \exp(-a(t_i - t_{i-1})) + b(1 - \exp(-a(t_i - t_{i-1}))) + \frac{Z}{\sqrt{2a}} \sqrt{\sigma^2 (1 - \exp(-2a(t_i - t_{i-1})))}
\quad \text{where} \quad Z \sim \mathcal{N}(0, 1)$

```python
def VasicekNextMargin(M, a, b, sigma, dt=1/12):     
    val1 = np.exp(-1*a*dt)
    val2 = (sigma**2)*(1-val1**2) / (2*a)
    out = M*val1 + b*(1-val1) + (np.sqrt(val2))*np.random.normal()
    return out
```

```python
def VasicekSim(N, M0, a, b, sigma, dt = 1/12):
    short_M = [0]*N  
    short_M[0] = M0     
    
    for i in range(1,N):
        short_M[i] = VasicekNextMargin(short_M[i-1], a, b, sigma, dt)
    
    return short_M
```

I use the above function to simulate the margin path. I fix a seed to have reproducibility in my results.

```python
years = 4
N = years * 12
t = np.arange(0,N)/12
np.random.seed(123)

test_sim = VasicekSim(N, M0, a, b, sigma, 1/12)

plt.figure(figsize=(20,6))
plt.plot(t,test_sim)
plt.xlabel('Time')
plt.ylabel('Values in USD/bbl')
plt.title('Future Refinery Margin based on the Vasicek model')
plt.show()
```

To evaluate the asset, I use Bachelier's model and the below formula:

$C = (F_0 - K)N\left(d\right) + \left(\sigma \sqrt{T - t}\right)N'\left(d\right)$
where $d = \frac{F_0 - K}{\sigma \sqrt{T - t}}$

```python
import numpy as np
from scipy.stats import norm

risk_free_rate = 0.05  
strike_price = 0 # we want a positive margin so the strike is 0

def bachelier_call_option_price(margin, volatility, time_to_expiry):
    forward_price = margin
    d1 = (forward_price - strike_price) / (volatility * np.sqrt(time_to_expiry))
    option_price = (forward_price - strike_price) * norm.cdf(d1) + volatility * np.sqrt(time_to_expiry) * norm.pdf(d1)
    return option_price

start_date = pd.to_datetime('2024-06-01')
end_date = pd.to_datetime('2027-12-31')

relevant_prices = Margin_vasicek[(Margin_vasicek.index >= start_date) & (Margin_vasicek.index <= end_date)]

positive_margin_dates = relevant_prices[relevant_prices['Future Margin Vasicek'] > 0].index.tolist()
expiry_date = relevant_prices.index[-1]

# Calculate option prices and discount to present value

total_option_value = 0
for date in positive_margin_dates:
    time_to_expiry = (expiry_date-date).days / 365
    margin_value = relevant_prices.at[date, 'Future Margin Vasicek']
    margin_volatility = sigma  # I use the sigma estimated from the Vasicek model
    option_price = bachelier_call_option_price(margin_value, margin_volatility, time_to_expiry)
    present_value = option_price * np.exp(-risk_free_rate * time_to_expiry) #discounting
    total_option_value += present_value

print("Value of the asset using Bachelier's model:", total_option_value*100000)
```
